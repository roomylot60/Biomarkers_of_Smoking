```python
col = ['정확도', '정밀도', '재현율', 'F1 score', 'AUC score']
idx = ['기본 모델링', '최적 하이퍼파라미터', 'MinMaxScaler', 'StandardScaler', 'RobustScaler','logScaler','이상치 제거']
data = [[0.8294,0.7565,0.7919,0.7738,0.9129],
        [0.9476,0.9265,0.9315,0.9290,0.9890],
        [0.9501,0.9287,0.9364,0.9325,0.9900],
        [0.9545,0.9351,0.9417,0.9384,0.9909],
        [0.9522,0.9312,0.9398,0.9355,0.9903],
        [0.9521,0.9274,0.9437,0.9355,0.9904],
        [0.9844,0.9758,0.9867,0.9812,0.9981]]
```

```python
result = pd.DataFrame(columns=col,index=idx)
result
```

```python
result.loc['기본 모델링'] = data[0]
result.loc['최적 하이퍼파라미터'] = data[1]
result.loc['MinMaxScaler'] = data[2]
result.loc['StandardScaler'] = data[3]
result.loc['RobustScaler'] = data[4]
result.loc['logScaler'] = data[5]
result.loc['이상치 제거'] = data[6]
result
```

```python
def bar_graph(col_name):
    result[col_name].plot(kind='bar',color = '#4999B6')
    for i in range(len(idx)):
        if result[col_name][i]==result[col_name].max():
            plt.text(i-0.25,result[col_name][i],round(result[col_name][i],3),color = 'red')

        else:
            plt.text(i-0.25,result[col_name][i],round(result[col_name][i],3))
    plt.xticks(rotation=70)
    plt.yticks(np.arange(0,1,0.3))
    plt.title('모델별'+col_name)
```

```python
bar_graph('정확도')
```

```python
def bar_graph_model(num,num2,color):
    result.iloc[num].plot(kind='bar',color = color)
    for i in range(0, 5):
            plt.text(i-0.25,result.iloc[num][i],round(result.iloc[num][i],3))
#     plt.figure(figsize=(12,10))
    plt.xticks(rotation=45)
    plt.yticks(np.arange(0,1,0.3))
    plt.title(result.index[num]+'&'+result.index[num2])
    plt.legend()
```

```python
bar_graph_model(1,1,'#4999B6')
bar_graph_model(0,1,'#FEFDA3')
```

```python
bar_graph_model(2,2,'#4999B6')
bar_graph_model(0,2,'#FEFDA3')
```
